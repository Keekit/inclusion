```python
# Pour les questions a choix multiple (dites binaires), nous allons attribué une equivalence en points
# Définition de la correspondance points / réponses
points_mapping = {
    'Jamais': 10,
    'Rarement': 8,
    'Parfois': 6,
    'Souvent': 4,
    'Régulièrement': 0
}

# Création de nouvelles colonnes pour les points attribués
for question in questions_binaires:
    new_column_name = f"{question} - Points"
    data[new_column_name] = data[question].map(points_mapping)

# Calcul de la moyenne des colonnes
binaire_moyenne = 'Moyenne des Points Binaires'
data[binaire_moyenne] = data[[f"{question} - Points" for question in questions_binaires]].mean(axis=1).round(2)
```
